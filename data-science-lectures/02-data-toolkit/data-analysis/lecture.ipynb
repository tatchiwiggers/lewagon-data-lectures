## NumPy
Numpy is a Fundamental package for high-performance data manipulation with Python

### What’s the difference between a Python list and a NumPy array?
NumPy gives you an enormous range of fast and efficient ways of creating arrays and manipulating numerical data inside them. While a Python list can contain different data types within a single list, all of the elements in a NumPy array should be homogeneous. The mathematical operations that are meant to be performed on arrays would be extremely inefficient if the arrays weren’t homogeneous.

### Why use NumPy?

NumPy arrays are faster and more compact than Python lists. An array consumes less memory and is more convenient to use. NumPy uses much less memory to store data and it provides a mechanism of specifying the data types. This allows the code to be optimized even further.

```python
# VANILLA PYTHON
# importing system module
import sys
  
S = range(1000)
print("Size of each element of python list in bytes: ", sys.getsizeof(S))
print("Size of the whole python list in bytes: ", sys.getsizeof(S)*len(S))
```

```python
# NUMPY
import numpy as np
import sys

D = np.arange(1000)
print("Size of each element of the Numpy array in bytes: ", D.itemsize)
print("Size of the whole Numpy array in bytes: ", D.size*D.itemsize)
```

```python
import numpy
import time

# size of arrays and lists
size = 1_000_000

# declaring python lists
list1 = range(size)
list2 = range(size)

# declaring numpy arrays
array1 = numpy.arange(size)
array2 = numpy.arange(size)

# capturing time before the multiplication of Python lists
initialTime = time.time()

# multiplying elements of both the lists and stored in another list
resultantList = [(a * b) for a, b in zip(list1, list2)]

# calculating execution time for the above operation
print("Time taken by Python Lists to perform multiplication:",
	(time.time() - initialTime),
	"seconds")

# capturing time before the multiplication of Numpy arrays
initialTime = time.time()

# multiplying elements of both the Numpy arrays and stored in another Numpy array
resultantArray = array1 * array2

# calculating execution time
print("Time taken by NumPy Arrays to perform multiplication:",
	(time.time() - initialTime),
	"seconds")
```

## Vectorized operations ⚡️

Okay. So now let's think how we can go into summing the contents of my array.<br>
These are vectorized operations.<br>
So we have a two dimensional array with eight rows, two columns and what we want to do from here is to create a one dimensional vector.

### Let's compute the sum row-by-row (8 additions) to create a 1D vector

```python
my_list = [
    [6, 5],
    [1, 3],
    [5, 6],
    [1, 4],
    [3, 7],
    [5, 8],
    [3, 5],
    [8, 4],
]
```

if we think about Python, what we would do is to again compute this operation using a list comprehension. It seems to have worked.

```python
[row[0] + row[1] for row in my_list]
```

But in numpy the logic is a little bit different.<br>
So we start with a two dimensional array with two columns in this case. And what numpy does is- it splits them into two separate one dimensional arrays, adds them together and gives you back one single dimensional array.<br>
So we're still working in the vertical space.<br>

So what I will do is to first create an array from my list and then say, Get me all the rows and the first column, add to it all the rows and the value from the second column. It will be stored in a new variable, which is my sum, which will be a new array.

```python
A = np.array(my_list)
my_sum = A[:,0] + A[:,1] # Vectorial "+" operator
my_sum
```

## Axes

Axes will come in handy when we start performing different operations such as data processing and the more dimensions you have, the more axes you get.<br>
When we start with one dimensional array, you just have one dimension and it will be characterized by the index zero. So it's the axis zero and the shape that you get back is a number in a tuple.<br>
And the second field is empty because you don't have any other dimensions.<br>
When we go to a two dimensional array, the axis zero becomes the one that goes down the rows. So we think vertically - axis one refers to the columns, so it goes horizontally. So if you want to perform some operation on just one column or on all the columns, you would say, I want to do it along axis zero.<br>
If you want to do something for a for example, on a particular row, you want to go horizontally. So you would say axis one, the shape goes into two numbers in a tuple. So the first one is number of rows. The other one is the number of columns. And when we go into the three dimensional array, we have axis zero going down the rows, axis one, going along the columns and axis two goes into the depth. So as you can see, the result of shape is also bigger. And we start always with the axis zero axis one, axis two - with what is the number of rows or columns and the particular dimension.

### Pandas vs Numpy
- Pandas is more user-friendly, but NumPy is faster. 
- Pandas has a lot more options for handling missing data, but NumPy has `better performance` on large datasets.
- Pandas uses Python objects internally, making it easier to work with NumPy (which uses C arrays)
- As it turns out, Pandas and NumPy libraries are similar in many ways and can be used interchangeably
- In my experience, Pandas is more powerful for data analysis.
- And though Pandas has only been around for a short time, it has been adopted by many developers and analysts since it was built to overcome the data analysis problems that many programmers had when using NumPy.
- NumPy is the core component of scientific computing in Python, while Pandas is more `useful for analyzing large datasets`. Both are powerful in their own right and are usually used together for large datasets. Based on your experience, you can definitely determine what is the best library for your use case.

## Pandas - Grouping

Very close to GROUP BY in SQL. It's a 3-step process:

- Split: a DataFrame is split into groups depending on keys
- Apply: an aggregative function (sum, mean, etc.) is applied on each group
- Combine: result from the previous operations are merged (i.e. reduced) into one new DataFrame

## 🤔 Which region of the world is the most populated?

```python
regions = countries_df.groupby('Region')
```

```python
regions[['Population', 'Area (sq. mi.)']].sum()
```

```python
regions[['Population', 'Area (sq. mi.)']].sum() \
    .sort_values('Population', ascending=False)
```
