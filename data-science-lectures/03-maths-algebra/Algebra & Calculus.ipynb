### Functions
Uma função é uma relação entre conjuntos que associa cada elemento do primeiro conjunto a exatamente **um elemento** do segundo conjunto. No primeiro set pegamos um valor, aplicamos esse valor à função e chegamos em um novo set com o resultado. A ideia é aqui é combinar o elemento do primeiro set com o elemento do segundo - não podemos ter dois valores para um mesmo elemento no primeiro set - importante lembrarmos disso.



Então aqui temos a função f que vai dos números reais aos números reais e a definição dessa função nos diz que para todo x o resultado da aplicação dessa função será 2x + 1. Aqui temos a notação matemática. 
Exemplo: 
\begin{equation*} f:R→R,f(x)=2x+1\end{equation*}

Em Python, implementamos esta função com um parâmetro (int), retornando um valor (int):

```python
def square_plus_one(x):
    return x * x + 1
```

De agora em diante vamos falar da notação f(x) como y.

#### Multi-variate functions
Uma função pode ser definida por mais de duas variáveis, como a função f que vai dos números reais aos números reais e a definição dessa função nos diz que para todo x0 e x1 o resultado da aplicação dessa função será 2x0 + 3x1 + 7:
\begin{equation*} f:R2→R,f(x0,x1)=2x0+3x1+7\end{equation*}

que será traduzido no mundo Python como:

```python
def multi_variate(x0, x1):
    y = 2 * x0 + 3 * x1 + 7
    return y
```

### Function representation
Quando temos uma função matématica, é sempre interessante plotar essa função no sistema de coordenadas e basicamente quando plotamos uma linha reta no plano, precisamos somente de dois números pra fazer essa plotagem.

Então, como falei, cada linha reta é definida por dois números na equação \begin{equation*} y = m x + b\end{equation*} onde a "m" se refere à inclinação, que é o famoso **slope** e "b" é o **intercept**, ou seja, o valor que corta o ponto "y" quando m for = 0.

No gráfico abaixo temos funções lineares plotadas onde o intercept é 0, ou seja, todos passam pela origem do gráfico. 

Temos a linha vermelha com a inclinação de 1/2, a linha verde tem a inclinação = 1 e a azul que tem a inclinação a 2.

Da pra perceber o quão rápido nossa inclinação aumenta ao longo do tempo baseado no valor de m, quanto maior o m, maior a inclinação da reta. Podemos mover o "b" também, atribuindo o valor 1 por exemplo. Esse gráfico moveria um ponto pra cima.

Ao longo da aula de hoje e nos modulos iniciais de machine learning, nós vamos trabalhar com modelos que traçam uma linha ou até mesmo um plano se estivermos falando de mais de 2 dimensões para conseguir descrever da melhor maneira possivel o comportmento das nossas colunas em relação ao nosso target (Y). Então, trazendo isso pra linguagem de machine learning, vamos desenhar uma reta ou um plano que consegue descrever da melhor maneira possivel o comportamento de nossos features (X), que são as colunas que preveem o target (Y) em machine learning de features.

Vamos então lembrar dessa equação porque vamos utilizá-la muito durante o curso.

```python
def linear_func(m, b, x):
    y = m * x + b
    return y
```

```python
# quando o x for = 2 teremos apenas u ponto porque o x é um escalar
x = 2
y = linear_func(1, 1, x)
plt.ylim(-100, 100)
plt.scatter(x, y)
plt.plot(x, y)
```

```python
# quando x for um vetor, como por exemplo com valores variando entre -10 e 10
# de dois em dois teremos uma reta. vai ser sempre uma reta
x = np.arange(-10, 10, 2)
y = linear_func(1, 1, x)
plt.ylim(-100, 100)
plt.scatter(x, y)
plt.plot(x, y)
```

```python
# se começarmos a modificar os slopes, vejam o que acontece:
slopes = [1, 10, 100]  #slope é o m
# x = [np.arange(-10, 10, 2) for slope in slopes]
for slope in slopes:
    x = np.arange(-10, 10, 2)
    y = linear_func(slope, 1, x)
    plt.ylim(-100, 100)
    plt.scatter([x], [y])
    plt.plot(x, y, label=f'slope {slope}')
    plt.legend()
# aqui testamos com 3 slopes diferentes, ou seja, o que vai mudar aqui é a inclinação da reta, certo
```

```python
# e o mesmo vale para o intercept, porem ele funciona de maneira diferente
# ele some soma ao produto do slope e da variável, por isso ele sempre vai dar
# um valor de partida
# por isso se começamos com o intercept = 1, o produto de mx vai ser negativo a principio
# mas se passamos por exemplo o valor de 100, noss reta vai começar em cima, no positivo
# conslusão, o intercept modifica a posição da reta no plano cartesiano
intercepts = [1, 10, 100]  #slope é o m
for intercept in intercepts:
    x = np.arange(-10, 10, 2)
    y = linear_func(1, intercept, x)
    #     plt.ylim(-100, 100)
    plt.scatter([x], [y])
    plt.plot(x, y, label=f'slope {slope}')
    plt.legend()
```

### Identity Function

A **função identidade**, também conhecida como **função inclusão** é aquela que tem a equação x = y, o que significa que nenhum valor é alterado. Então, se x = 1, y = 1 também. A **função identidade** nos dá uma linha reta exatamente a 45 graus dividindo o plano ao meio -  então importante lembrar que a identidade é algo que quando você aplica um valor a x, x não muda:
\begin{equation*} f(x) = x\end{equation*}

### Inverse Function

É uma função que "inverte" outra função: se a função f aplicada a uma entrada x dá um resultado de y, então aplicando sua função inversa g a y dá o resultado x, e vice-versa.

\begin{equation*} f(x)=y⟺g(y)=x\end{equation*}

⚠️ Nem todas as funções têm uma inversa. Mas caso exista, g é frequentemente denotado como f^−1 e ambos f e f^−1 são bijeções. 

Em matemática, uma bijeção, também conhecida como função bijetiva, é a correspondência um-para-um, onde cada elemento de um conjunto é emparelhado com exatamente um elemento do outro conjunto e cada elemento do outro conjunto é emparelhado com exatamente um elemento do primeiro conjunto.

Isso significa que g de f é a identidade porque quando pegamos g(f(x)) = x

# System of linear equations

Um sistema de equações lineares (ou sistema linear) é uma coleção de uma ou mais equações lineares envolvendo o mesmo conjunto de variáveis.

- 3x+2y−z = 1
- 2x−2y+4z = −2
- −x+1/2y−z = 0

Uma solução para um sistema linear é uma atribuição de valores às variáveis de modo que todas as equações sejam satisfeitas **simultaneamente**. Então a ideia aqui é buscar uma solução onde basicamente o resultado de x, y e z será válido se, se somente se, quando verificamos todas as equações, constatamos que elas são verdadeiras com os valores dos resultados x, y, z. Por exemplo:

- x = 1
- y = −2
- z = −2

A própria palavra sistema indica que as equações devem ser consideradas coletivamente, e não individualmente.



### Trivial example (one unknown) 
2x = 4

Existem sistemas de 1, 2, 3 4 quantas varíáveis quisermos. Se o sistema tiver uma variável, teremos uma equação. Nesse caso a equação é bem simpes, todos concordam que x = 2? Como que chegamos nesse valor? "Passandoo dois pra la né?" Mas não é bem assim tá? A principal aqui é deixar essa equação igual, ou seja, balanceada, o que quer dizer que o que fizermos em um lado da equação precisa ser feito do outro, e o objetivo é isolar o x e para isso temos a fórmula matematica que seria os dois lados da equação por dois resultando em x = 2.

### Simple non-trivial example (two unknowns)
🤔 Problema:

- Um pai é 22 anos mais velho que seu filho. 
- Em 10 anos, o pai terá o dobro da idade do filho. 
- Quantos anos eles tem?

Vamos colocar esse problema em uma equação e resolvê-la. Como fazer isso? Podemos dizer então que y será a idade do pai e x será a idade do filho.

- y = father e x = son
- y = x + 22
- y + 10 = 2(x + 10) => porque o filho será 10 anos mais velho tambem quando se passarem 10 anos.


```python
# Reescrevendo:
y = x + 22
y = 2 * x + 10
```

```python
# podemos isolar a idade do pai porque dessa forma podemos
# testar valores diferentes para a idade do filho
x = np.arange(6, 18)
x
plt.plot(x, x + 22)
plt.plot(x, 2 * x + 10)
plt.xlabel('son')
plt.ylabel('father')
plt.show()
# então fazendo o cruzamento entre as duas retas, conseguimos encontrar
# a idade do filho e a idade do pai
```

### Euclidian Distance

 
 coordenadas = A(2, 4), B=(3, 2)
 
 A distância euclidiana é a distância entre dois pontos, que pode ser provada pela aplicação do teorema de Pitágoras.
 Temos duas coordenadas onde o 1o elemento é sempre referente a X e o segundo a Y.
 Conectado esses pontos e passasndo duas linhas pelo eixo X e Y, criamos um triangulo - aogra podemos calcular os valores de 
 nossos catetos.
 - 4 - 2 = 2
 - 3 - 2 = 1

```python
x = [2, 3]
y = [4, 2]

plt.scatter(x, y)
plt.ylim(0, 5)
plt.xlim(0, 4)
```

```python
#  Agora podemos calcular a hipotenusa
xa = x[0]
xb = x[1]

ya = y[0]
yb = y[1]

import math

distancia = math.sqrt((xa - xb)**2 + (ya - yb)**2)
#distancia = ((xa - xb) ** 2 + (ya - yb) ** 2) ** 0.5 # tira a raiz quadrada
distancia
```
