### Functions

A ideia √© aqui √© combinar o elemento do primeiro set com o elemento do segundo - n√£o podemos ter dois valores para um mesmo elemento no primeiro set - importante lembrarmos disso.



Ent√£o aqui temos a fun√ß√£o f que vai dos n√∫meros reais aos n√∫meros reais e a defini√ß√£o dessa fun√ß√£o nos diz que para todo x o resultado da aplica√ß√£o dessa fun√ß√£o ser√° 2x + 1. Aqui temos a nota√ß√£o matem√°tica.

#### Multi-variate functions
Uma fun√ß√£o pode ser definida por mais de duas vari√°veis, como a fun√ß√£o f que vai dos n√∫meros reais aos n√∫meros reais e a defini√ß√£o dessa fun√ß√£o nos diz que para todo x0 e x1 o resultado da aplica√ß√£o dessa fun√ß√£o ser√° 2x0 + 3x1 + 7:
\begin{equation*} f:R2‚ÜíR,f(x0,x1)=2x0+3x1+7\end{equation*}

que ser√° traduzido no mundo Python como:

```python
def multi_variate(x0, x1):
    y = 2 * x0 + 3 * x1 + 7
    return y
```

### Function representation
Quando temos uma fun√ß√£o mat√©matica, √© sempre interessante plotar essa fun√ß√£o no sistema de coordenadas e basicamente quando plotamos uma linha reta no plano, precisamos somente de dois n√∫meros pra fazer essa plotagem.

Ent√£o, como falei, cada linha reta √© definida por dois n√∫meros na equa√ß√£o \begin{equation*} y = m x + b\end{equation*} onde a "m" se refere √† inclina√ß√£o, que √© o famoso **slope** e "b" √© o **intercept**, ou seja, o valor que corta o ponto "y" quando m for = 0.

No gr√°fico abaixo temos fun√ß√µes lineares plotadas onde o intercept √© 0, ou seja, todos passam pela origem do gr√°fico. 

Temos a linha vermelha com a inclina√ß√£o de 1/2, a linha verde tem a inclina√ß√£o = 1 e a azul que tem a inclina√ß√£o a 2.

Da pra perceber o qu√£o r√°pido nossa inclina√ß√£o aumenta ao longo do tempo baseado no valor de m, quanto maior o m, maior a inclina√ß√£o da reta. Podemos mover o "b" tamb√©m, atribuindo o valor 1 por exemplo. Esse gr√°fico moveria um ponto pra cima.

Ao longo da aula de hoje e nos modulos iniciais de machine learning, n√≥s vamos trabalhar com modelos que tra√ßam uma linha ou at√© mesmo um plano se estivermos falando de mais de 2 dimens√µes para conseguir descrever da melhor maneira possivel o comportmento das nossas colunas em rela√ß√£o ao nosso target (Y). Ent√£o, trazendo isso pra linguagem de machine learning, vamos desenhar uma reta ou um plano que consegue descrever da melhor maneira possivel o comportamento de nossos features (X), que s√£o as colunas que preveem o target (Y) em machine learning de features.

Vamos ent√£o lembrar dessa equa√ß√£o porque vamos utiliz√°-la muito durante o curso.

```python
def linear_func(m, b, x):
    y = m * x + b
    return y
```

```python
# quando o x for = 2 teremos apenas u ponto porque o x √© um escalar
x = 2
y = linear_func(1, 1, x)
plt.ylim(-100, 100)
plt.scatter(x, y)
plt.plot(x, y)
```

```python
# quando x for um vetor, como por exemplo com valores variando entre -10 e 10
# de dois em dois teremos uma reta. vai ser sempre uma reta
x = np.arange(-10, 10, 2)
y = linear_func(1, 1, x)
plt.ylim(-100, 100)
plt.scatter(x, y)
plt.plot(x, y)
```

```python
# se come√ßarmos a modificar os slopes, vejam o que acontece:
slopes = [1, 10, 100]  #slope √© o m
# x = [np.arange(-10, 10, 2) for slope in slopes]
for slope in slopes:
    x = np.arange(-10, 10, 2)
    y = linear_func(slope, 1, x)
    plt.ylim(-100, 100)
    plt.scatter([x], [y])
    plt.plot(x, y, label=f'slope {slope}')
    plt.legend()
# aqui testamos com 3 slopes diferentes, ou seja, o que vai mudar aqui √© a inclina√ß√£o da reta, certo
```

```python
# e o mesmo vale para o intercept, porem ele funciona de maneira diferente
# ele some soma ao produto do slope e da vari√°vel, por isso ele sempre vai dar
# um valor de partida
# por isso se come√ßamos com o intercept = 1, o produto de mx vai ser negativo a principio
# mas se passamos por exemplo o valor de 100, noss reta vai come√ßar em cima, no positivo
# conslus√£o, o intercept modifica a posi√ß√£o da reta no plano cartesiano
intercepts = [1, 10, 100]  #slope √© o m
for intercept in intercepts:
    x = np.arange(-10, 10, 2)
    y = linear_func(1, intercept, x)
    #     plt.ylim(-100, 100)
    plt.scatter([x], [y])
    plt.plot(x, y, label=f'slope {slope}')
    plt.legend()
```

### Identity Function

A **fun√ß√£o identidade**, tamb√©m conhecida como **fun√ß√£o inclus√£o** √© aquela que tem a equa√ß√£o x = y, o que significa que nenhum valor √© alterado. Ent√£o, se x = 1, y = 1 tamb√©m. A **fun√ß√£o identidade** nos d√° uma linha reta exatamente a 45 graus dividindo o plano ao meio -  ent√£o importante lembrar que a identidade √© algo que quando voc√™ aplica um valor a x, x n√£o muda:
\begin{equation*} f(x) = x\end{equation*}

### Inverse Function

√â uma fun√ß√£o que "inverte" outra fun√ß√£o: se a fun√ß√£o f aplicada a uma entrada x d√° um resultado de y, ent√£o aplicando sua fun√ß√£o inversa g a y d√° o resultado x, e vice-versa.

\begin{equation*} f(x)=y‚ü∫g(y)=x\end{equation*}

‚ö†Ô∏è Nem todas as fun√ß√µes t√™m uma inversa. Mas caso exista, g √© frequentemente denotado como f^‚àí1 e ambos f e f^‚àí1 s√£o bije√ß√µes. 

Em matem√°tica, uma bije√ß√£o, tamb√©m conhecida como fun√ß√£o bijetiva, √© a correspond√™ncia um-para-um, onde cada elemento de um conjunto √© emparelhado com exatamente um elemento do outro conjunto e cada elemento do outro conjunto √© emparelhado com exatamente um elemento do primeiro conjunto.

Isso significa que g de f √© a identidade porque quando pegamos g(f(x)) = x

# System of linear equations

Um sistema de equa√ß√µes lineares (ou sistema linear) √© uma cole√ß√£o de uma ou mais equa√ß√µes lineares envolvendo o mesmo conjunto de vari√°veis.

- 3x+2y‚àíz = 1
- 2x‚àí2y+4z = ‚àí2
- ‚àíx+1/2y‚àíz = 0

Uma solu√ß√£o para um sistema linear √© uma atribui√ß√£o de valores √†s vari√°veis de modo que todas as equa√ß√µes sejam satisfeitas **simultaneamente**. Ent√£o a ideia aqui √© buscar uma solu√ß√£o onde basicamente o resultado de x, y e z ser√° v√°lido se, se somente se, quando verificamos todas as equa√ß√µes, constatamos que elas s√£o verdadeiras com os valores dos resultados x, y, z. Por exemplo:

- x = 1
- y = ‚àí2
- z = ‚àí2

A pr√≥pria palavra sistema indica que as equa√ß√µes devem ser consideradas coletivamente, e n√£o individualmente.



### Trivial example (one unknown) 
2x = 4

Existem sistemas de 1, 2, 3 4 quantas var√≠√°veis quisermos. Se o sistema tiver uma vari√°vel, teremos uma equa√ß√£o. Nesse caso a equa√ß√£o √© bem simpes, todos concordam que x = 2? Como que chegamos nesse valor? "Passandoo dois pra la n√©?" Mas n√£o √© bem assim t√°? A principal aqui √© deixar essa equa√ß√£o igual, ou seja, balanceada, o que quer dizer que o que fizermos em um lado da equa√ß√£o precisa ser feito do outro, e o objetivo √© isolar o x e para isso temos a f√≥rmula matematica que seria os dois lados da equa√ß√£o por dois resultando em x = 2.

### Simple non-trivial example (two unknowns)
ü§î Problema:

- Um pai √© 22 anos mais velho que seu filho. 
- Em 10 anos, o pai ter√° o dobro da idade do filho. 
- Quantos anos eles tem?

Vamos colocar esse problema em uma equa√ß√£o e resolv√™-la. Como fazer isso? Podemos dizer ent√£o que y ser√° a idade do pai e x ser√° a idade do filho.

- y = father e x = son
- y = x + 22
- y + 10 = 2(x + 10) => porque o filho ser√° 10 anos mais velho tambem quando se passarem 10 anos.


```python
# Reescrevendo:
y = x + 22
y = 2 * x + 10
```

```python
# podemos isolar a idade do pai porque dessa forma podemos
# testar valores diferentes para a idade do filho
x = np.arange(6, 18)
x
plt.plot(x, x + 22)
plt.plot(x, 2 * x + 10)
plt.xlabel('son')
plt.ylabel('father')
plt.show()
# ent√£o fazendo o cruzamento entre as duas retas, conseguimos encontrar
# a idade do filho e a idade do pai
```

### Euclidian Distance

 
 coordenadas = A(2, 4), B=(3, 2)
 
 A dist√¢ncia euclidiana √© a dist√¢ncia entre dois pontos, que pode ser provada pela aplica√ß√£o do teorema de Pit√°goras.
 Temos duas coordenadas onde o 1o elemento √© sempre referente a X e o segundo a Y.
 Conectado esses pontos e passasndo duas linhas pelo eixo X e Y, criamos um triangulo - aogra podemos calcular os valores de 
 nossos catetos.
 - 4 - 2 = 2
 - 3 - 2 = 1

```python
x = [2, 3]
y = [4, 2]

plt.scatter(x, y)
plt.ylim(0, 5)
plt.xlim(0, 4)
```

```python
#  Agora podemos calcular a hipotenusa
xa = x[0]
xb = x[1]

ya = y[0]
yb = y[1]

import math

distancia = math.sqrt((xa - xb)**2 + (ya - yb)**2)
#distancia = ((xa - xb) ** 2 + (ya - yb) ** 2) ** 0.5 # tira a raiz quadrada
distancia
```

```python
def plot_line(a, b, ax):
    """Plot a line from slope and intercept on the ax"""

    x_vals = np.array(ax.get_xlim())
    y_vals = a * x_vals + b
    ax.plot(x_vals, y_vals, '-')

```
