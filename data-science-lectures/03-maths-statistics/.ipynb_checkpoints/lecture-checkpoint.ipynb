## Statistical dispersion - calculates the dispersion of our data
Dispersion (also called variability, scatter, or spread) is the extent to which a distribution is stretched or squeezed.

Examples:

- Variance œÉ^2
- Standard deviation œÉ -  is a measure of the amount of variation or dispersion of a set of values. 
- Interquatile Range IQR - When a dataset has outliers, variability is often summarized by a statistic called the interquartile range, which is the difference between the first and third quartiles. 

## Variance

variance is a measure of dispersion that takes into account the spread of all data points in a dataset.

- **Variance** - (sum of values - mean)^2  divided by the number of elements N.
- **Standard deviation** - is the square root of the (sum of values - mean)^2  divided by the number of elements N.
- **Sample** - is the (sum of values - mean)^2 divided by the number of elements N - 1. We are only working with a sample, not with the entire universe of elements.

## Interquartile range (IQR)

**If a value stands outside the IQR, whether it be less than Q1 - 1.5 x IQR or greater than Q3 + 1.5 x IQR - that value is an outlier.**

## Random experiment
A random experiment is a process by which we observe something uncertain. After the experiment, the result of the random experiment is known: it is the outcome.

The set of all possible outcomes is called the sample space S, Œ© or U.

Examples:

- Toss a coin. S=H,T
- Roll a dice. S=1,2,3,4,5,6
- Observe the number of goals in a soccer match. S=0,1,2,3,‚Ä¶

When we repeat a random experiment several times, we call each one of them a trial.

**Sample space** is defined based on how you define the random experiment.

If the experiment is Toss a coin three times then the sample space is: 2^3

- S={(H,H,H),(H,H,T),(H,T,H),(T,H,H),(H,T,T),(T,H,T),(T,T,H),(T,T,T)}
- The goal is to assign a probability to events, defined as subsets of a sample space S.

## Probability
We assign a probability measure P(A) to an event A.

- This is a value between 0 and 1 that shows how likely the event is.
- Some property you can easily visualize with venn diagrams:

### Conditional Probability

#### Example
Let's consider the random experiment: draw two cards, one at a time, without replacement, in a deck of 52 cards.

You draw the first card: it's a King (event A).<br>
‚ùì What's the probability that the second card is also a King (event B)?

```python
# Solution
# The first card is not placed back in the deck which means A and B are dependant.
prob_a = 4 / 52
prob_b_given_a = 3 / 53 # this is based on the fact that a card was already taken from the deck
prob_a_and_b = prob_a * prob_b_given_a # here getting 2kings in a row(P(A) and P(B))
prob_a_and_b
```

## Baye's Theorem


```python
# P(Sick / +)=?
# P(+‚à£ Healhy)=P(+‚à£ Sick^‚àÅ)=0.02
prob_healthy = 0.999
prob_pos_given_healthy = 0.02

prob_sick = 0.001
prob_pos_given_sick = 0.99

prob_sick_and_pos = prob_sick * prob_pos_given_sick
prob_healthy_and_pos = prob_healthy * prob_pos_given_healthy

prob_pos = prob_sick_and_pos + prob_healthy_and_pos

prob_sick_given_pos = (prob_pos_given_sick * prob_sick) / prob_pos
prob_sick_given_pos * 100
```

## Normal Distribution
√â uma curva normal dada por um mean(Œº) a variance(œÉ¬≤):<br>


```python
def plot_normal_distribution(mu, variance):
    sigma = math.sqrt(variance)
    x = np.linspace(-10, 10, 100)
    plt.plot(x, statsb.norm.pdf(x, mu, sigma), label=f"Œº={mu}, œÉ¬≤={variance}")

plot_normal_distribution(0, 1)
plot_normal_distribution(1, 2)
plot_normal_distribution(-3, 5)
plt.legend()
plt.show()
```

```python
def plot_cumulative_normal_distribution(mu, variance):
    sigma = math.sqrt(variance)
    x = np.linspace(-10, 10, 100)
    plt.plot(x, statsb.norm.cdf(x, mu, sigma), label=f"Œº={mu}, œÉ¬≤={variance}")

plot_cumulative_normal_distribution(0, 1)
plot_cumulative_normal_distribution(1, 2)
plot_cumulative_normal_distribution(-3, 5)
plt.legend()
plt.show()
```
